```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import tensorflow as tf
from tensorflow import keras
```

```python
df = pd.read_csv('cust_churn1.csv')
df=df.drop('Unnamed: 0',axis=1)
```

```python
x = df.drop('Churn',axis=1)
y = df['Churn']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=10)
```

```python
# now using deeplearning model
```

```python
model= keras.Sequential([
    keras.layers.Dense(20, input_shape=(26,),activation='relu'),
    keras.layers.Dense(20,activation='relu'),
    keras.layers.Dense(1,activation='sigmoid')
    
])
model.compile(optimizer='adam',
             loss= 'binary_crossentropy',
             metrics=['accuracy'])
```

1st layer is an input layer, in which the number of neurons is 20, no. of columns in input, activation function is relu
the second one is hidden layer.
the third layer is output layer in which the activation function is sigmoid
the loss function is binary cross entropy as out output is a binary running 5 epochs firsy

```python
model.fit(x_train, y_train,epochs=100)
```

here we are getting 83 percent of accuracy, now we will evaluate the model on test data

```python
model.evaluate(x_test,y_test)
```

accuracy on testing data is nearly 80% so our model is working good on test data too

now we are predicting the model

```python
y_p=model.predict(x_test)
```

```python
y_p[:10]
```

```python
# we will convert the 2d array to 1d array and then convert all the values to 0 or 1
```

```python
y_test[:10]
```

```python
y_pred = []
for i in y_p:
    if i>0.5:
        y_pred.append(1)
    else:
        y_pred.append(0)
```

```python
y_pred[:10]
```

as it is 80% accuracy so, it makes some mistakes

now we are making the classification report for this model

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
print(classification_report(y_test,y_pred))
```

```python
# making confusion matrix and its heatmap
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
# This line creates a confusion matrix using TensorFlow's built-in confusion_matrix function
cm = tf.math.confusion_matrix(labels=y_test,predictions=y_pred)
# heatmap of the confusion matrix cm using the heatmap function from the seaborn visualization library
#  fmt='d' tells seaborn to format these annotations as integers
# The annot=True argument tells seaborn to annotate each cell of the heatmap with the value of the corresponding element in the confusion matrix
plt.figure(figsize=(10,7))
sns.heatmap(cm,annot=True,fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

the term "annotate" typically refers to the process of adding labels or annotations to a dataset. fmt='d' to ensure that the annotations are formatted as integers rather than as decimals or scientific notation.
